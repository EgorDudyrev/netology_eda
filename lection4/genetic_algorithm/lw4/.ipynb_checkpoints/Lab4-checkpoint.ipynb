# Define functions

```python
def generate_connections(n_servers, con_p=0.5, min_length=10, max_length=50):
    """
    Create random connections in graph
    """
    connections = {}
    for i in range(n_servers):
        for j in range(i+1, n_servers):
            if np.random.rand()<con_p:
                v = np.random.randint(min_length, max_length+1)
                connections[(i,j)] = v
    return connections
```

```python
def generate_matrix(connection_dict, n_servers):
    """
    Create graph matrix based on given connections
    """
    A = np.ones((n_servers,n_servers)) * np.inf
    for i in range(n_servers):
        A[i,i]=0
    for coord, val in connection_dict.items():
        A[coord]=val
        A[coord[::-1]] = val
    return A
```

```python
def plot_graph(G, start_node, end_node, best_edges=None):
    """
    G is a Graph from networkx library
    """
    colors = ['green' if i==start_node else 'red' if i==end_node else 'lightgrey' for i in range(len(G.nodes))]
    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, node_color=colors, width=1)
    labels = nx.get_edge_attributes(G,'weight')
    labels = {k:int(v) for k,v in labels.items()}
    if best_edges:
        best_edges_w = {e:int(A[e]) for e in best_edges}
        nx.draw(G, pos, with_labels=True, node_color=colors, edgelist=best_edges, width=3, edge_color='g')
        nx.draw_networkx_edge_labels(G, pos,edge_labels=best_edges_w)
    else:
        nx.draw_networkx_edge_labels(G, pos,edge_labels=labels)
    plt.show()
```

```python
def generate_path(path_len, start, stop, n_servers):
    """
    Create random path. first and last elements are fixed 'start' and 'stop'
    """
    return np.hstack([start, np.random.randint(0, n_servers, path_len-2), stop])

def get_path_len(path):
    return sum([A[a,b] for a,b in zip(path, path[1:])])

def get_fitness(path):
    """
    Main idea: the more path_lenth - the less fitness
    """
    return 1/get_path_len(path)

def select_pairs(paths):
    return np.array([(paths[i], paths[j]) for i,j in np.random.randint(0, len(paths), size=(N//2, 2))])

def crossover(a, b):
    return np.array([i if np.random.rand()<0.5 else j for i,j in zip(a,b)])

def mutation(path):
    mutated = np.array([i if np.random.rand()<0.9 else np.random.randint(N_SERVERS) for i in path[1:-1]])
    return np.hstack([path[0], mutated, path[-1]])
```

```python
def select_manually(paths):
    to_print = [(i, get_path_len(path), path,) for i,path in enumerate(paths)]
    to_print = sorted(to_print, key = lambda line: line[1])
    ids_map = {line[0]:idx for idx, line in enumerate(to_print)}
    ids_map_r = {v:k for k,v in ids_map.items()}
    clear_output()
    print('Select best lines')
    for line in to_print:
        print("{id_}: {l}, (len={length})".format(id_=ids_map[line[0]], l=line[2], length=line[1]))
    print('Enter best lines ids:')
    ids = input().strip()
    if ids=='stop':
        raise StopIteration
    if len(ids)>0:
        ids = list(map(int, ids.split(' ')))
        ids = [ids_map_r[i] for i in ids]
        parents = paths[ids]
    else:
        #parents = np.array([generate_path(len(paths[0]), paths[0][0], paths[0][-1], N_SERVERS) for i in range(N)])
        parents = paths
    print('-----------------\n')
    return parents
```

```python
def ga_step(mx, paths, is_manual=False):
    if len(paths)%2==1: raise Exception('Paths number should be odd')
    # step2: calculate fitness
    fits = np.array([get_fitness(path) for path in paths])
    fits = np.nan_to_num(fits)/np.nan_to_num(fits.mean()+1e-10)
    # step3: select parents vectors
    if is_manual:
        parents = select_manually(paths)
    else:
        probs = np.exp(fits)/np.exp(fits).sum() #Softmax function
        parents = np.array([paths[i] for i in np.random.choice(range(len(paths)), p=probs, size=len(paths))])
    
    # step4: set pairs. apply crossover and mutation
    pairs = select_pairs(parents)
    children = np.array([crossover(pair[0],pair[1]) for pair in pairs])
    children = np.vstack([children, children])
    children = np.array([mutation(child) for child in children])
    return children
```

# Create graph

```python
N_SERVERS = 10
start_node, end_node = 0, 9
```

```python
connection_dict = generate_connections(N_SERVERS, con_p=0.7)
```

```python
A = generate_matrix(connection_dict, N_SERVERS) #Matrix to work with
G = nx.Graph(A) #Graph object for pretty plots
edges_to_remove = [ (i,j) for i in range(N_SERVERS) for j in range(N_SERVERS) if np.isinf(A[i,j])]
G.remove_edges_from(edges_to_remove)
```

# Manual optimization

```python
N=10
path_len = 10
is_manual = True
```

```python
paths = np.array([generate_path(path_len, start_node, end_node, N_SERVERS) for i in range(N)])
all_paths = np.array([paths])
stats = np.array([min([get_path_len(i) for i in paths])])
```

```python
plot_graph(G, start_node, end_node)
```

```python
plot_data = np.array(stats)
plot_data[np.isinf(plot_data)]=plot_data[~np.isinf(plot_data)].max()*1.2 #Replace inf with finite numbers
```
