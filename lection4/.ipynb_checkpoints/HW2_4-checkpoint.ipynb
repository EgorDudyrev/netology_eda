### TRDATETIME

Дата транзакции

Вычленим из TRDATETIME информацию о дне, месяце и т.п.

```python
ds['TRDATETIME_dt'] = pd.to_datetime(ds['TRDATETIME'], format='%d%b%y:%H:%M:%S')
```

```python
ds['day'] = [x.day for x in ds['TRDATETIME_dt']]
ds['month'] = [x.month for x in ds['TRDATETIME_dt']]
ds['year'] = [x.year for x in ds['TRDATETIME_dt']]
ds['weekday'] = [x.weekday() for x in ds['TRDATETIME_dt']]
```

```python
ds['hour'] = [x.hour for x in ds['TRDATETIME_dt']]
```

```python
ds['is_holiday'] = (ds['weekday']>=5).astype(int)
```

```python
ds['is_holiday'].value_counts()
```

```python
plt.figure(figsize=(15,8))
for idx, f in enumerate(['day','month','year','weekday','hour','is_holiday']):
    plt.subplot(2,3,idx+1)
    s = ds.pivot_table(index=f, columns='target_flag', values='cl_id', aggfunc='nunique')
    s = s/s.sum()
    sns.heatmap(s.T, cmap='Blues')
    plt.title(f)
plt.tight_layout()
plt.show()
```

Кажется нет никакой разницы между простыми признаками времени

Кажется, что клиент должен переставать пользоваться услугами банка постепенно. Т.е. его частота транзакций должна увеличиваться во времени.

Посмотрим как целевая переменная зависит от средней периодичности транзакций клиента

```python
dss = []
for cl in tqdm_notebook(ds['cl_id'].drop_duplicates()):
    cds = ds[ds['cl_id']==cl].sort_values('TRDATETIME_dt')
    s = cds['TRDATETIME_dt']
    s = s-s.shift(1)
    s = [x.days for x in s]
    cds['days_since_last_trans'] = s
    dss.append(cds)
ds = pd.concat(dss)
ds['days_since_last_trans'] = ds['days_since_last_trans'].fillna(0)
```

```python
cds = ds.groupby('cl_id')[['days_since_last_trans','target_flag']].mean()
sns.distplot(np.log1p(cds[cds['target_flag']==0]['days_since_last_trans']), label='target False')
sns.distplot(np.log1p(cds[cds['target_flag']==1]['days_since_last_trans']), label='target True')
plt.legend()
plt.show()
```

Особой разницы в распределениях не видно

Можно посмотреть изменение периодичности транзакций во времени: как часто клиент совершил свои последние 1,2,3,...,10 транзакций относительно своей средней частоты

```python
dss = []
for cl in tqdm_notebook(ds['cl_id'].drop_duplicates()):
    cds = ds[ds['cl_id']==cl].sort_values('TRDATETIME_dt')
    s = cds['days_since_last_trans']
    for i in range(1,11):
        cds[f'last_{i}_trans_freq'] = np.mean(s[-i:])
        cds[f'last_{i}_trans_coef'] = np.mean(s[-i:])/np.mean(s) if np.mean(s)!=0 else 0
    dss.append(cds)
ds = pd.concat(dss)
```

```python
ds.loc[:,ds.drop('MCC_group',1).columns] = ds.drop('MCC_group',1).fillna(0)
```

```python
plt.figure(figsize=(15,8))
for i in range(1,11):
    plt.subplot(4,3,i)
    cds = ds.groupby('cl_id')[[f'last_{i}_trans_freq','target_flag']].mean()
    sns.distplot(np.log1p(cds[cds['target_flag']==0][f'last_{i}_trans_freq']), label='target False')
    sns.distplot(np.log1p(cds[cds['target_flag']==1][f'last_{i}_trans_freq']), label='target True')
    plt.title(f'last {i}')
    plt.legend()
plt.tight_layout()
plt.suptitle('last transactions frequency')
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
plt.figure(figsize=(15,8))
for i in range(1,11):
    plt.subplot(4,3,i)
    cds = ds.groupby('cl_id')[[f'last_{i}_trans_coef','target_flag']].mean()
    sns.distplot(np.log1p(cds[cds['target_flag']==0][f'last_{i}_trans_coef']), label='target False')
    sns.distplot(np.log1p(cds[cds['target_flag']==1][f'last_{i}_trans_coef']), label='target True')
    plt.title(f'last {i}')
    plt.legend()
plt.tight_layout()
plt.suptitle('last transactions coefs')
plt.subplots_adjust(top=0.9)
plt.show()
```

Наибольшее разделение кажется достигается при сравнении частоты последних 5 транзакций со всеми

## Построение модели

```python
np.random.seed(42)
test_clients = np.sort(np.random.choice(ds['cl_id'].drop_duplicates(), size=1000, replace=False))
train_clients = ds[-ds['cl_id'].isin(test_clients)]['cl_id'].drop_duplicates().sort_values().values
```

## Предсказание на исходных признаках

```python
y_feat = 'target_flag'
real_feats = ['amount']
cat_feats = ['PERIOD','MCC_desc', 'channel_type', 'currency','trx_category']
bin_feats = [x for x in cat_feats if ds[x].nunique()<=2]
cat_feats = list(set(cat_feats)-set(bin_feats))
```

```python
dss = []
for f in cat_feats:
    cds = ds.pivot_table(index='cl_id', columns=f, values='amount', aggfunc='mean', fill_value=0)
    cds.columns = [f+'__'+str(x) for x in cds.columns]
    dss.append(cds)
cds = pd.concat(dss,1)
for f in real_feats+bin_feats:
    cds[f] = ds.groupby('cl_id')[f].mean()
cds['target_flag'] = ds.groupby('cl_id')['target_flag'].mean()
```

```python
X,Y = cds.drop(y_feat,1), cds[y_feat]

idx = X.index.isin(train_clients)
X_train, Y_train = X[idx], Y[idx]

idx = X.index.isin(test_clients)
X_test, Y_test = X[idx], Y[idx]
```

```python
xgb = XGBClassifier(seed=42)
xgb.fit(X_train, Y_train)
preds = xgb.predict_proba(X_test)[:,1]
```

```python
score = roc_auc_score(Y_test, preds)
fprs, tprs, treshs = roc_curve(Y_test, preds)
plt.plot(fprs, tprs, label=f'score: {score:.4f}')
plt.plot([0,1],[0,1], label='score: 0.5')
plt.title(f'ROC curve')
plt.legend()
plt.show()
```

```python
feat_imps = pd.Series(xgb.feature_importances_, index=X_train.columns)
feat_imps = feat_imps[feat_imps>0]
feat_imps = feat_imps[feat_imps>0.01*feat_imps.sum()].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(y=feat_imps.index, x=feat_imps.values)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def get_XY(ds, cat_feats, real_feats, bin_feats, train_clients=train_clients, test_clients=test_clients):
    dss = []
    for f in cat_feats:
        cds = ds.pivot_table(index='cl_id', columns=f, values='amount', aggfunc='mean', fill_value=0)
        cds.columns = [f+'__'+str(x) for x in cds.columns]
        dss.append(cds)
    cds = pd.concat(dss,1)
    for f in real_feats+bin_feats:
        cds[f] = ds.groupby('cl_id')[f].mean()
    cds['target_flag'] = ds.groupby('cl_id')['target_flag'].mean()
    
    X,Y = cds.drop(y_feat,1), cds[y_feat]

    idx = X.index.isin(train_clients)
    X_train, Y_train = X[idx], Y[idx]

    idx = X.index.isin(test_clients)
    X_test, Y_test = X[idx], Y[idx]
    return X_train, Y_train, X_test, Y_test
```

```python
def plot_roc(Y_test_list, preds_list):
    Y_test_list = Y_test_list if type(Y_test_list)==list else [Y_test_list]
    preds_list = preds_list if type(preds_list)==list else [preds_list]
    for idx, zipped in enumerate(zip(Y_test_list, preds_list)):
        Y_test, preds = zipped
        fprs, tprs, treshs = roc_curve(Y_test, preds)
        plt.plot(fprs, tprs, label=f'score {idx}: {score:.4f}')
    plt.plot([0,1],[0,1], label='score: 0.5')
    plt.title(f'ROC curve')
    plt.legend()
```

```python
def calc_importances(rf, features):
    fimps = pd.DataFrame()
    fimps['feature'] = features
    fimps['value'] = rf.feature_importances_
    fimps = fimps.sort_values('value')
    fimps.index = range(len(fimps))
    fimps['feature_base'] = [x if '__' not in x else x.split('__')[0] for x in fimps['feature']]
    d = fimps.groupby('feature_base')['value'].median().to_dict()
    fimps['value_median'] = fimps['feature_base'].apply(lambda x: d[x])
    d = fimps.groupby('feature_base')['value'].mean().to_dict()
    fimps['value_mean'] = fimps['feature_base'].apply(lambda x: d[x])
    res = d.copy()
    res['roc_auc'] = score
    return fimps, res
```

Дополнительно возмём синус от периодических переменных (час, день, месяц), чтобы показать, чтобы использовать их в качестве числовых признаков (синус покажет, что 00:00 ближе к 23:00, чем к 2:00).

```python
for f in ['hour','day','month','year','weekday']:
    min_, max_ = ds[f].min(), ds[f].max()
    ds[f+'_sin'] = np.sin((ds[f]-min_)/(max_-min_+1)*2*np.pi)
```

```python
for idx,f in enumerate(['hour','day','month','year','weekday']):
    plt.subplot(2,3,idx+1)
    sns.distplot(ds[f+'_sin'])
    plt.title(f+'_sin')
plt.tight_layout()
plt.show()
```

Полный перебор займёт слишком много времени

Когда-то в университет я проходил генетические алгоритмы

Данная задача в терминах генетических алгоритмов:
* gene - столбец, признак
* chromo - набор столбцов
* population - несколько вариантов столбцов

Попробуем найти лучшую комбинацию из 30 признаков

```python
def generate_chromo(columns):
    cols = sorted(np.random.choice(columns, size=CHROMO_LEN, replace=False))
    return cols
def get_model(chromo):
    X_train, Y_train, X_test, Y_test = get_XY(ds, cat_feats, real_feats, bin_feats)
    rf = RandomForestClassifier(n_estimators=50, random_state=42)
    rf.fit(X_train[chromo], Y_train)
    rf.feature_names_ = chromo
    return rf
def get_fitness(model):
    preds = model.predict_proba(X_test[model.feature_names_])[:,1]
    score = roc_auc_score(Y_test, preds)
    return score
def select_parents(population, fits):
    probs = np.exp(fits)/np.exp(fits).sum() #Softmax function
    parents = [population[i] for i in np.random.choice(range(POPULATION_LEN),
                                                       p=probs, size=POPULATION_LEN)]
    return parents
def select_pairs(parents):
    return [(parents[i], parents[j]) for i,j in
                   np.random.randint(0, POPULATION_LEN, size=(POPULATION_LEN//2, 2))]
def crossover(pair):
    a,b = pair
    feats_ds = pd.DataFrame()
    feats_ds['feature'] = a.feature_names_+b.feature_names_
    feats_ds['value'] = list(a.feature_importances_)+list(b.feature_importances_)

    feats_ds = feats_ds.groupby('feature')['value'].mean()
    if len(feats_ds)<CHROMO_LEN:
        return a.feature_names_
    feats_ds = np.exp(feats_ds)/np.exp(feats_ds).sum()
    return sorted(np.random.choice(feats_ds.index, CHROMO_LEN, p=feats_ds.values, replace=False))
def mutation(chromo):
    mutated = [gene if np.random.rand()<1-MUTATION_P
                        else np.random.choice(list(set(X_train.columns)-set(gene)))
                        for gene in chromo]
    return sorted(mutated)
def reduction(old_population, children, fits=None):
    old_population_ids = list(range(len(old_population)))
    old_population_fits_ar = [(i, f) for i,f in zip(old_population_ids, fits)]
    old_population_ids = [i for i,f in sorted(old_population_fits_ar, key=lambda x: -x[1])]
    old_population = [old_population[i] for i in range(len(old_population)) if i in old_population_ids]
    tresh = max(int(POPULATION_LEN/10),1)
    new_population = old_population[:tresh]+children[:POPULATION_LEN-tresh]
    return new_population
```

```python
def ga_step(old_population):
    population = old_population.copy()
    models = [get_model(g) for g in population]
    scores = [get_fitness(m) for m in models]
    parents = select_parents(models, scores)
    pairs = select_pairs(parents)
    children = [crossover(pair) for pair in pairs+pairs] #Each pair gives 2 children
    mutated = [mutation(child) for child in children]
    new_population = reduction(population, mutated, scores)
                             
    return new_population, scores
```

```python
CHROMO_LEN = 70
POPULATION_LEN = 50
MUTATION_P = 0.1
N_EPOCHS = 1000
EARLY_STOPPING_STEPS = 50
```

```python
population = [generate_chromo(X_train.columns) for i in range(POPULATION_LEN)]
scores_total = []
max_scores = []
populations_total = []
for i in tqdm_notebook(range(N_EPOCHS)):
    try:
        new_population, scores = ga_step(population)
    except StopIteration:
        print("Stopped at",i)
        break
    
    
    scores_total.append(scores)
    populations_total.append(population)

    max_score = max(scores)
    max_scores.append(max_score)
    print(max_score)
    if i>=EARLY_STOPPING_STEPS \
        and all(max_score>=max_scores[- EARLY_STOPPING_STEPS:]):
        print('early stops at',i)
        print('result', max_score)
        break
    population = new_population
max_ind = np.argmax(scores_total)
best_chromo = populations_total[max_ind//POPULATION_LEN][max_ind%POPULATION_LEN]
```

```python
scores_ds = pd.DataFrame(scores_total)
scores_ds = np.concatenate([scores_ds[f] for f in scores_ds.columns])
scores_ds = pd.DataFrame(scores_ds, columns=['roc_auc'])
```
